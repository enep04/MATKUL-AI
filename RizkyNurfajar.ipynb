```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Membaca file CSV dari Google Drive
data = pd.read_csv('/content/drive/MyDrive/penyakit malaria/malaria_data_2020_dummy.csv')

# Menambah kolom 'Fatality Rate' dengan rumus (Deaths / Cases) * 100
data['Fatality Rate'] = (data['Deaths'] / data['Cases']) * 100

# Menampilkan data dengan Fatality Rate
print(data[['Country', 'Region', 'Year', 'Cases', 'Deaths', 'Fatality Rate']].head())

# Statistik Deskriptif untuk Analisis Risiko
fatality_stats = data['Fatality Rate'].describe()
print("\nStatistik Deskriptif Tingkat Kematian:\n", fatality_stats)

# Visualisasi Tingkat Risiko Kematian per Negara
plt.figure(figsize=(10, 6))
plt.bar(data['Country'], data['Fatality Rate'], color='salmon')
plt.title('Tingkat Risiko Kematian Akibat Malaria per Negara (2020)')
plt.xlabel('Negara')
plt.ylabel('Fatality Rate (%)')
plt.xticks(rotation=45)
plt.show()

# Visualisasi Risiko Kematian Rata-Rata per Wilayah
region_fatality = data.groupby('Region')['Fatality Rate'].mean()

plt.figure(figsize=(8, 5))
region_fatality.plot(kind='bar', color='teal')
plt.title('Rata-Rata Tingkat Risiko Kematian Akibat Malaria per Wilayah (2020)')
plt.xlabel('Wilayah')
plt.ylabel('Fatality Rate (%)')
plt.xticks(rotation=45)
plt.show()
```
